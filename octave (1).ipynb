## MODEL EVALUATION

```python
New_data['TAXI_OUT'].hist()
```

Taxi out time does not follow normal distribution,hence normalization or standardization

```python
#creating dependent and independent variables
x=New_data.loc[:,(['CRS_DEP_TIME', 'DAY_OF_MONTH', 'DAY_OF_WEEK', 'DEST', 'DISTANCE',
       'DISTANCE_GROUP', 'MONTH', 'OP_CARRIER', 'OP_CARRIER_AIRLINE_ID',
       'OP_CARRIER_FL_NUM', 'OP_UNIQUE_CARRIER', 'ORIGIN', 'ORIGIN_AIRPORT_ID',
       'ORIGIN_AIRPORT_SEQ_ID', 'ORIGIN_CITY_MARKET_ID', 'ORIGIN_CITY_NAME',
       'ORIGIN_STATE_FIPS', 'ORIGIN_STATE_NM', 'ORIGIN_WAC', 'QUARTER',
       'TAIL_NUM', 'TAXI_OUT'])].values
y=New_data.loc[:,'TAXI_OUT'].values
```

```python
New_data.columns
```

```python
from sklearn.preprocessing import MinMaxScaler #normalizing of the independent variable
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
print(x)
```

# Train-Test split

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression,BayesianRidge,HuberRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score,r2_score,mean_absolute_error,mean_squared_error,mean_squared_log_error
```

```python
linReg=LinearRegression()
linReg.fit(x_train,y_train)
y_pred=linReg.predict(x_test)
print('The r-squared is',r2_score(y_test,y_pred))
print('The mean absolute error is',mean_absolute_error(y_test,y_pred))
print('The mean squared error is',mean_squared_error(y_test,y_pred))
```

```python
bayesian=BayesianRidge()
bayesian.fit(x_train,y_train)
y_pred1=bayesian.predict(x_test)
print('The r-squared is',r2_score(y_test,y_pred1))
print('The mean absolute error is',mean_absolute_error(y_test,y_pred1))
print('The mean squared error is',mean_squared_error(y_test,y_pred1))
```

Linear Regresion model is the best model,hence should be used to predict  predicts the taxi-out time for an aircraft 2 hours before the scheduled departure time
